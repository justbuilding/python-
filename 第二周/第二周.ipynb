```python
#购物车系统
#jupyter notebook不能用exit()，会断开服务器，写项目还是用pycharm的python，不用ipython
product_list =[('iphonex 256G',8368),
              ('ThinkPad p7',63999),
              ('Macbook Pro',13488),
              ('珠江钢琴',3800000),
              ('Kingston 64Gb',99.9),]
shopping_list = []
salary = input("Please input your salary: ")
if salary.isdigit():
    salary = float(salary)
    while True:     
        for index,item in enumerate(product_list):
            print(index,item)
        user_choice = input("请选择你要买的商品(输入商品数字,退出输入q)：")
        if user_choice.isdigit():#isdiit判断输入的是不是整数数字
                  user_choice = int(user_choice)
                  if user_choice < len(product_list) and user_choice >=0:
                      p_item = product_list[user_choice]
                      if p_item[1] <= salary:
                            shopping_list.append(p_item)
                            salary -= p_item[1]
                            print("Added %s into shopping cart,your current balance is \033[31;1m%s\033[0m]"%(p_item,salary))
#\033[31;1m%s\033[0m]表示吧%s所指示的变为红色
                      else:
                          print("\033[41;1m]You only own %s ,you cannot afford it.\0333[0m"%salary)
                  else:
                    print("product [%s] is not exist"%user_choice)
        elif user_choice == 'q':
            print("-----------shopping list-----------")
            for p in shopping_list:
                print(p)
            print("Your current balance:",salary)
            #exit(0)
        else:
            print("invalid option")
    else:
        #exit(0)
else:
    print("You have quit")
    #exit()
```

```python
name = "hjj\taal"
print(name.capitalize())#首字母大写
print(name.count("a"))#计数
print(name.center(50,"-"))#包含name指定的数据和-总共50个，name居中显示
print(name.endswith("ex"))#是否以ex结尾
print(name.expandtabs(tabsize=30))#\t代表tab,输出tab时设置间隔4变为30，jupyter使用tab成空格，很少用到
print(name.find("j"))#找到第一个j在哪个index
print(name[name.find("jj"):])#通过find来截断字符串
name2 = "my name is {name} and I am {age}."
print(name2.format(name='hjj',age =23))#常用到
print(name2.format_map({'name':'hjj','age':23}))
print('123'.isalnum())#是否都是数字
print('absd'.isalpha())#是否都是字母
print('1'.isdecimal())#是否为十进制,不用于小数判断，decimal为“小数，十进制”
print('1'.isdigit())#判断是否为整数
print('a 1A'.isidentifier())#是否为合法标识符
print(' '.isspace())#是否为空格
print('My Name'.istitle())#是否为标题，判断依据是所有首字母有大写
print('MM'.isupper())#是否都为大写
print('+'.join(['1','2','3']))#将列表转换为字符串
print(name.ljust(30,'*'))#左对齐
print('hjj'.upper())#转为大写
print('hJJ'.lower())
print('\nhjj0'.lstrip())#去掉左边的\n
print('hjj1\n'.rstrip())
print('\nhjj1\n'.strip())#去掉两边
```

```python
p = str.maketrans("abcJQyli",'123$@456')
print("Jay Q".translate(p))#把"abcJQyli"和'123$@456'一一对应翻译
print('JayQ jay Q'.rfind('Q'))#会从左查找第一个Q,输出其的index
print('Jay Q jay q q q'.replace('q','Q',2))#把q替换成Q,只替换第一个和第二个共两个
print('1+2+3+4'.split('+'))#按照+把数字划分出来
print('1+2\n+3+4'.splitlines())#按照\n来划分，自动识别不同系统的换行符
#下面有例子
str1 = 'ab c\n\nde fg\rkl\r\n'
print(str1.splitlines())
str2 = 'ab c\n\nde fg\rkl\r\n'
print(str2.splitlines(True))

print('Jay Q jay q'.swapcase())#大小写替换
print('32'.zfill(30))
```

```python
#字典
info = {
    'stu1':"dkf",
    'stu2':"eoi",
    'stu3':"dki",
    'stu4':"iie"}
print(info)
b = {
    'stu9':"ii",
    "stu1":"iio"
}
#info添加b，有的元素则修改value，没有的元素就创建
info.update(b)
print(info)
#转换，把字典转换为列表
print(info.items())
#修改或增加，同一种方法
info['stu4']="ies"
info['stu5']="cci"
print(info)
#删除,三种方法
del info['stu2']
print(info)
info.pop("stu1")
print(info)
info.popitem()
print(info)
#查找
print(info.get('stu3'))
print('stu3' in info)
#嵌套
mm = {
    "111":{
        "aa":["1a"],

        "bb":["2b"],

        "cc":["3c"]
    },
    "444":{
        "dd":["4d"],

        "ee":["5e"],

        "ff":["6f"]
    },
    "777":{
        "gg":["7g"],

        "hh":["8h"],

        "jj":["9j"]
    }
}
mm["777"]["gg"][0] = "233"
print(mm)
mm.setdefault("777",{"kk":["eeee"]})#如果存在则不创建
mm.setdefault("888",{"kk":["eeee"]})#如果不存在则创建
print(mm)
```
